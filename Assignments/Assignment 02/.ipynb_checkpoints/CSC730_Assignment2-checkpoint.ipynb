## Fuzzy C-Means Implementation

```python
def calculate_objective_function(data, cluster_centers, membership_matrix, fuzziness):
    
    distances = np.linalg.norm(data[:, np.newaxis] - cluster_centers.T, axis=2)
    objective_function_value = np.sum(membership_matrix**fuzziness * distances**2)
    
    return objective_function_value
```

```python
def fuzzy_c_means(data, num_clusters, fuzziness, max_iter=100, tolerance=1e-4):
    # Initialize membership matrix randomly
    membership_matrix = np.random.rand(len(data), num_clusters)
    membership_matrix /= np.sum(membership_matrix, axis=1, keepdims=True)

    for _ in range(max_iter):
        # Update cluster centers
        cluster_centers = np.dot(data.T, membership_matrix**fuzziness) / np.sum(membership_matrix**fuzziness, axis=0, keepdims=True)

        # Calculate distances between data points and cluster centers
        distances = np.linalg.norm(data[:, np.newaxis] - cluster_centers.T, axis=2)

        # Update membership matrix
        new_membership_matrix = 1 / np.sum((distances[:, :, np.newaxis] / 
                                            distances[:, np.newaxis, :])**(2/(fuzziness-1)), axis=2)

        new_membership_matrix /= np.sum(new_membership_matrix, axis=1, keepdims=True)
        
        # Calculate the change in the objective function value
        obj_function_value = calculate_objective_function(data, cluster_centers, membership_matrix, fuzziness)
        new_obj_function_value = calculate_objective_function(data, cluster_centers, new_membership_matrix, fuzziness)
        obj_function_change = abs(new_obj_function_value - obj_function_value)

        # Check for convergence
        if obj_function_change < tolerance:
            break
            
        membership_matrix = new_membership_matrix

    return cluster_centers, membership_matrix

```

### 784 Dimensions

## Adjusted Rand Index (ARI)

```python
def ari_from_scratch(true_labels, pred_labels):
    # Create a contingency table from true and predicted labels
    contingency_table = contingency_matrix(true_labels, pred_labels)

    n_choose_k = np.vectorize(math.comb)

    # Calculate the total number of ways to choose 2 items from all elements
    n_choose_2 = n_choose_k(np.sum(contingency_table),2)
    
    # Sum of combinations for each cell in the contingency table
    n_choose_2_sum = np.sum(n_choose_k(contingency_table,2))

    # Calculate the sum of elements in each row and each column of the contingency table
    row_sum = np.sum(contingency_table,axis=1)
    col_sum = np.sum(contingency_table,axis=0)
    
    # Sum of combinations for each row and each column
    row_sum_choose_2 = np.sum(n_choose_k(row_sum,2))
    col_sum_choose_2 = np.sum(n_choose_k(col_sum,2))

    # Actual index: total number of pairings in the contingency table
    index = n_choose_2_sum
    
    # Calculate the expected index (chance grouping)
    expectedIndex = row_sum_choose_2 * (col_sum_choose_2/n_choose_2)
    
    # Maximum index: average of the sum of row combinations and column combinations
    maxIndex = (row_sum_choose_2 + col_sum_choose_2)/2
    
    ari = (index - expectedIndex)/(maxIndex - expectedIndex)
    return contingency_table, ari
```

```python
contingency_2, ari_2 = ari_from_scratch(y,clusters_2)
print("Contingency Table for 2 dimensions: \n", contingency_2)
print("ARI for 2 dimensions: ", ari_2)

print("ARI using scikit-learn:", adjusted_rand_score(y, clusters_2))
```

```python
contingency_784, ari_784 = ari_from_scratch(y,clusters_784)
print("Contingency Table for 784 dimensions: \n", contingency_784)
print("ARI for 784 dimensions: ", ari_784)

print("ARI using scikit-learn:", adjusted_rand_score(y, clusters_784))
```
